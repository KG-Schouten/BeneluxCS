```python
with open("C:/Python codes/BeneluxCS/DiscordBot/data/league_teams_open1_8.json", "r") as file:
    league_teams = json.load(file)

# Flatten members into a list of dictionaries
flattened_members = []
for team, data in league_teams.items():
    for member in data["members"]:
        member["team"] = team  # Add team name for context
        flattened_members.append(member)

df = pd.DataFrame(flattened_members)

# Ensure memberships column is always a list (replace NaN with an empty list)
df["memberships"] = df["memberships"].apply(lambda x: x if isinstance(x, list) else [])

# Filter: country == 'nl' and 'esea' in memberships
filtered_df = df[(df["country"].isin(["nl", "be", 'lu'])) & df["memberships"].apply(lambda x: "esea" in x)]

filtered_df["faceit_url"] = filtered_df["faceit_url"].str.replace("{lang}", "en")

# Display the results
print(filtered_df)
```
